1.	Agrupa los siguientes datos en intervalos de amplitud 2:
2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6
Construye una tabla de frecuencias completa y dibuja el histograma correspondiente.  



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos
datos = [2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1,
         1, 5, 6, 7, 4, 3, 2, 5, 4, 6]
amplitud = 2

# Intervalos
bins = list(range(min(datos), max(datos) + amplitud, amplitud))
fa, edges = np.histogram(datos, bins=bins)
fr = fa / len(datos)
F = np.cumsum(fa)

# Tabla
tabla = pd.DataFrame({
    "Intervalo": [f"[{edges[i]}, {edges[i+1]})" for i in range(len(edges)-1)],
    "f": fa,
    "fr": fr.round(2),
    "F": F
})
tabla.style.background_gradient(cmap="YlGnBu").set_caption("Tabla de Frecuencias - Ejercicio 1")

# Gr√°fico
plt.figure(figsize=(8,5))
palette = plt.cm.Set3.colors  # Paleta de 12 colores bonitos
color_random = palette[np.random.randint(len(palette))]
plt.hist(datos, bins=bins, color=color_random, edgecolor="black", alpha=0.85, rwidth=0.9)
plt.xticks(bins)
plt.title("Histograma - Ejercicio 1", fontsize=14, fontweight='bold')
plt.xlabel("Intervalos")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
colores_fav = {
    "Rojo": 35,
    "Azul": 25,
    "Verde": 20,
    "Amarillo": 10,
    "Negro": 10
}

# Tabla con √°ngulos
tabla_colores = pd.DataFrame(list(colores_fav.items()), columns=["Color", "Porcentaje"])
tabla_colores["√Ångulo"] = (tabla_colores["Porcentaje"] * 360 / 100).round(2)
tabla_colores.style.background_gradient(cmap="coolwarm").set_caption("Colores Favoritos - Ejercicio 17")

# Gr√°fico
plt.figure(figsize=(7,7))
colores = plt.cm.tab20(range(len(colores_fav)))
plt.pie(tabla_colores["Porcentaje"], labels=tabla_colores["Color"],
        autopct='%1.1f%%', startangle=90, colors=colores, shadow=True)
plt.title("Diagrama de Sectores - Ejercicio 17", fontsize=14, fontweight='bold')
plt.show()
```

18.	En una encuesta sobre las frutas preferidas por un grupo de personas, se obtuvieron los siguientes datos:
o	Manzana: 50% o 	Banana: 30%
o	Naranja: 20%
a)	Representa estos datos mediante un diagrama de sectores.
b)	Calcula el √°ngulo correspondiente a cada sector.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
frutas = {
    "Manzana": 50,
    "Banana": 30,
    "Naranja": 20
}

# Tabla con √°ngulos
tabla_frutas = pd.DataFrame(list(frutas.items()), columns=["Fruta", "Porcentaje"])
tabla_frutas["√Ångulo"] = (tabla_frutas["Porcentaje"] * 360 / 100).round(2)
tabla_frutas.style.background_gradient(cmap="YlOrRd").set_caption("Frutas Preferidas - Ejercicio 18")

# Gr√°fico
plt.figure(figsize=(7,7))
colores = plt.cm.Set2(range(len(frutas)))
plt.pie(tabla_frutas["Porcentaje"], labels=tabla_frutas["Fruta"],
        autopct='%1.1f%%', startangle=90, colors=colores, shadow=True)
plt.title("Diagrama de Sectores - Ejercicio 18", fontsize=14, fontweight='bold')
plt.show()
```

19.	La distribuci√≥n del gasto mensual de una familia en diferentes categor√≠as es la siguiente:

o	Otros: 5%
a)	Representa estos datos mediante un diagrama de sectores.
b)	Calcula el √°ngulo correspondiente a cada sector.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
gasto_mensual = {
    "Alquiler": 40,
    "Alimentaci√≥n": 30,
    "Transporte": 15,
    "Ocio": 10,
    "Otros": 5
}

# Tabla con √°ngulos
tabla_gasto_mensual = pd.DataFrame(list(gasto_mensual.items()), columns=["Categor√≠a", "Porcentaje"])
tabla_gasto_mensual["√Ångulo"] = (tabla_gasto_mensual["Porcentaje"] * 360 / 100).round(2)
tabla_gasto_mensual.style.background_gradient(cmap="PuBuGn").set_caption("Gasto Mensual - Ejercicio 19")

# Gr√°fico
plt.figure(figsize=(7,7))
colores = plt.cm.Dark2(range(len(gasto_mensual)))
plt.pie(tabla_gasto_mensual["Porcentaje"], labels=tabla_gasto_mensual["Categor√≠a"],
        autopct='%1.1f%%', startangle=90, colors=colores, shadow=True)
plt.title("Diagrama de Sectores - Ejercicio 19", fontsize=14, fontweight='bold')
plt.show()
```
